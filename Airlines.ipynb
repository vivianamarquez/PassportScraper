```python
import pandas as pd
import bs4 as bs
import urllib.request
import random
import time
```

```python
test = 'https://www.airlinequality.com/review-pages/a-z-airline-reviews/'
```

```python
request = urllib.request.Request(test, headers={'User-Agent': 'Mozilla/5.0'})

webpage = urllib.request.urlopen(request)
source = webpage.read()
webpage.close()

soup = bs.BeautifulSoup(source, 'html.parser')
```

```python
airlines = []

for e in soup.find_all(class_='items'):
    for link in e.find_all('a', href=True):
        url = f"https://www.airlinequality.com{link['href']}"
        airlines.append(url)
```

```python
scores = []
names = []
review_counts = []

for airline in airlines:
    pause = random.randrange(100) / 100.0
    time.sleep(pause)
    
    request = urllib.request.Request(airline, headers={'User-Agent': 'Mozilla/5.0'})

    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()

    soup = bs.BeautifulSoup(source, 'html.parser')
    
    try:
        score = soup.find_all('span', itemprop='ratingValue')[0].text.strip()
        review_count = soup.find_all('span', itemprop='reviewCount')[0].text.strip()
        name = soup.find('h1').text.strip()
    except:
        score = None
        review_count = None
        name = None
        
    scores.append(score)
    names.append(name)
    review_counts.append(review_count)
    
    print(f"{name} has {score}/10 with {review_count} reviews.")
```

```python
df = [airlines, scores, names, review_counts]
df = pd.DataFrame(df).T
df.columns = ['links', 'score', 'name', 'reviews']
```

```python
df.score.value_counts()
```

```python
df.score = df.score.apply(lambda val: int(val))
df.reviews = df.reviews.apply(lambda val: int(val))

df[df.reviews>30].sort_values(['score', 'reviews'], ascending=False).head(30)
```
